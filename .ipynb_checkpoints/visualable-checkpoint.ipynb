```python
import matplotlib.pyplot as plt
import os
import numpy as np
from keras.preprocessing.image import img_to_array, load_img
from tqdm import tqdm
```

```python
def exp_pic_size(path):
    heights, widths = [], []
    for filenames in tqdm(os.listdir(path)):
        img = load_img(os.path.join(path, filenames))
        x = img_to_array(img) # this is a Numpy array with shape (n_c, n_w, n_h)
        heights.append(x.shape[0])
        widths.append(x.shape[1])
        
    x = np.array(widths)
    y = np.array(heights)
    
    plt.scatter(x, y, c='r', alpha=0.1, marker = 'o')
    plt.title('the size of pictures')
    plt.xlabel('width')
    plt.ylabel('height')
    plt.show()
```

```python
path = os.getcwd()
path_cat = os.path.join(path, r'train\\cat')
exp_pic_size(path_cat)
```

```python
path_dog = os.path.join(path, r'train\\dog')
exp_pic_size(path_dog)
```

```python
path_test = os.path.join(path, r'test')
exp_pic_size(path_test)
```
